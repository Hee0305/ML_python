```python
import pandas as pd
import numpy as np
```

```python
# Eaxmple Data
raw_data = {'first_name': ['Jason', np.nan, 'Tina', 'Jake', 'Amy'],
        'last_name': ['Miller', np.nan, 'Ali', 'Milner', 'Cooze'],
        'age': [42, np.nan, 36, 24, 73],
        'sex': ['m', np.nan, 'f', 'm', 'f'],
        'preTestScore': [4, np.nan, np.nan, 2, 3],
        'postTestScore': [25, np.nan, np.nan, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'sex', 'preTestScore', 'postTestScore'])
df
```

```python
df.isnull().sum()
df.isnull().sum() / len(df) # % 로 표기 
```

```python
df_cleaned = df.dropna(how='all') # 모든 데이터가 비어 있는 데이터를 제거 
df_cleaned = df.dropna(axis=1, how='all') # 칼럼을 기준으로, 모든 데이터가 비어 있는 데이터를 제거 
df_cleaned
```

```python
df_cleaned = df.dropna(axis=1, thresh=3) # 데이터가 4개 이상 없을 때 제거 
df_cleaned
```

```python
df.fillna(0) # Nan 값에 0 을 넣어라 
```

```python
# 데이터를 정규분포로 생각하였을때 = 평균값, 중위값, 최빈값을 활용 

df['preTestScore'].mean() # 평균값
df['preTestScore'].median() # 중위값
df['preTestScore'].mode() # 최빈값
 
df["preTestScore"].fillna(df["preTestScore"].mean(), inplace=True) # NaN 값에 평균값을 채워라 
df
```

```python
# 비어져 있는 곳이 성별값을 기준으로 평균값으로 채워짐 
df["postTestScore"].fillna(df.groupby("sex")["postTestScore"].transform("mean"), inplace=True)
df
```

```python
# make DataFrame
edges = pd.DataFrame({'source':[0,1,2],
                     'target':[2,2,3],
                     'weight':[3,4,5],
                     'color':['red','blue','blue']})
edges
```

```python
# pandas를 이용한 one-hot encoding

pd.get_dummies(edges)

# 필요한 데이터만 끊어서 one-hot encoding

pd.get_dummies(edges["color"])
```

```python
# 내신과 같이 데이터 내에서 등급을 나눈 것들은 값을 지정해 줄 수 있다. 

weight_dict = {3:'M',4:'L',5:'XL'} # weight 의 3,4,5 를 M,L,XL로 weight_sign이라는 column을 추가 
edges["weight_sign"] = edges["weight"].map(weight_dict)
edges
```

```python
# make DataFrame

raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
        'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'],
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['regiment', 'company', 'name', 'preTestScore', 'postTestScore'])
df
```

```python
# Binning

bins=[0, 25, 50, 75, 100] # Define bins as 0~25, 25~50, 60~75, 75~100
group_names = ['Low','Okay','Good','Great']
categories = pd.cut(df['postTestScore'], bins, labels=group_names)
categories
```

```python
# 위에서 만든 categories 를 기존 df에 추가 

df['categories'] = pd.cut(df['postTestScore'], bins, labels=group_names)
df
```

```python
# Feature scaling with sklearn 
# - Lamel encoder와 마찬가지로, sklearn도 feature scale 지원
# - Min-MaxScaler와 StandardScaler 사용 

from sklearn import preprocessing 

std_scale = preprocessing.StandardScaler().fit(df[['preTestScore', 'postTestScore']]) # fitting 과정
df_std = std_scale.transform(df[['preTestScore', 'postTestScore']]) # 적용 과정 
df_std[:5]
```
