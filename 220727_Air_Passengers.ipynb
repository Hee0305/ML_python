```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('./AirPassengers.csv')
print(df)
```

```python
df["step"] = range(len(df))
df.head()
```

```python
df["cum_sum"] = df["#Passengers"].cumsum()
df["cum_max"] = df["#Passengers"].cummax()
df["cum_min"] = df["#Passengers"].cummin()

df.head()
```

```python
import numpy as np
temp_data = df["Month"].map(lambda x:x.split('-'))
temp_date = np.array(temp_data.values.tolist())
temp_date
```

```python
df["year"] = temp_date[:,0]
df["month"] = temp_date[:,1]
df
```

```python
df["diff"] = df["#Passengers"].diff().fillna(0)
df.head()
```

```python
df["#Passengers"].pct_change()
```
